```python
import s3prl.hub as hub
import pickle
import os
from pydub import AudioSegment
import glob
import torch
import numpy as np
import re
from natsort import natsorted
import natsort
```

```python
def extract_audio_features(audio_path, model_name):
  device = 'cpu'
  waveform = AudioSegment.from_wav(audio_path)
  waveform = waveform.set_channels(1)
  waveform = waveform.get_array_of_samples()
  waveform = (torch.tensor(waveform, dtype = torch.float)).to(device)
  model =  getattr(hub, model_name)()
  model.num_layers=12
  # or cpu
  model= model.to(device)
  wavs = [waveform]
  with torch.no_grad():
      reps = model(wavs)["hidden_states"]
  return reps
```

```python
def extract_audio_features_context(audio_path1, audio_path2, model_name):
  device = 'cpu'
  waveform1 = AudioSegment.from_wav(audio_path1)
  waveform2 = AudioSegment.from_wav(audio_path2)
  waveform = waveform1 + waveform2
  waveform = waveform.set_channels(1)
  waveform = waveform.get_array_of_samples()
  waveform = (torch.tensor(waveform, dtype = torch.float)).to(device)
  model =  getattr(hub, model_name)()
  model.num_layers=12
  # or cpu
  model= model.to(device)
  wavs = [waveform]
  with torch.no_grad():
      reps = model(wavs)["hidden_states"]
  return reps
```

```python
def extract_audio_features_context2(audio_path1, audio_path2,audio_path3, model_name):
  device = 'cpu'
  waveform1 = AudioSegment.from_wav(audio_path1)
  waveform2 = AudioSegment.from_wav(audio_path2)
  waveform3 = AudioSegment.from_wav(audio_path3)
  waveform = waveform1 + waveform2 + waveform3
  waveform = waveform.set_channels(1)
  waveform = waveform.get_array_of_samples()
  waveform = (torch.tensor(waveform, dtype = torch.float)).to(device)
  model =  getattr(hub, model_name)()
  model.num_layers=12
  # or cpu
  model= model.to(device)
  wavs = [waveform]
  with torch.no_grad():
      reps = model(wavs)["hidden_states"]
  return reps
```
