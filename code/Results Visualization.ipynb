```python
def load_results(exp, kind, info, nchan, ntime, concat=False, reshape=True):
    if exp in ["context_audio", "context_text", "audio", "bert", "pengi_text"]:
        files = sorted(glob.glob(f"/media/srijith/easystore/srijith/pred/meg_sub*_predictions/{exp}_joint_embeddings_predictions/"))
    else:
        files = sorted(glob.glob(f"../../Research/pred/meg_sub*_predictions/{exp}_joint_embeddings_predictions/"))
    
    if kind in ["corr", "r2s"]:
        if exp in ["context_audio", "context_text", "audio", "bert", "pengi_text"]:
            res_fp = sorted(glob.glob(f"../pred/meg_sub*_predictions/{exp}_joint_embeddings_predictions/0_{kind}.npy"))
        else:
            res_fp = sorted(glob.glob(f"../../Research/pred/meg_sub*_predictions/{exp}_joint_embeddings_predictions/0_{kind}.npy"))
    #     res_fp = sorted(glob.glob(f"../reports/sub*/{exp}_s0_predictions/*{kind}*"))
    
    print(f"Loading {exp}, found sub: {len(files)}, found files: {len(res_fp)}")

    preds = dict()
    for fp in tqdm(files):
        sub = fp.split("/")[-3].split("-")[-1]
        
        if kind in ["corr", "r2s"]:
            res_files = sorted(glob.glob(f"{fp}/0_{kind}.npy"))
            try:
                if reshape:
                    preds[sub] = np.load(res_files[0]).reshape(-1, nchan, ntime).squeeze()
                else:
                     preds[sub] = np.load(res_files[0])
            except:
                print(f"Missing: {exp} - {kind}, Sub: {sub} ")
                continue
                
        else:
            res_files = sorted(glob.glob(f"{fp}/0_{kind}*"))
            results = []
            for story in range(4):
                if reshape:
                    results.append(np.load(res_files[story]).reshape(-1, nchan, ntime).squeeze())
                else:
                    results.append(np.load(res_files[story]))
            preds[sub] = np.concatenate(results)
        
    if concat:
        preds = np.concatenate(list(preds.values()))
    return preds


def as_evoked(*data, concat=False):
    evoked = []
    for datum in data:
        if concat:
            pred_epochs = np.concatenate(list(datum.values()))
            # print(pred_epochs.shape)
            evoked.append(mne.EpochsArray(pred_epochs, raw.info).average())
        else:
            pred_epochs = dict()
            for sub, preds in datum.items():
                pred_epochs[sub] = mne.EpochsArray(preds, raw.info).average()
                # print(pred_epochs.shape)
            evoked.append(pred_epochs)
        
    if len(evoked) > 1:
        return tuple(evoked)
    else:
        return evoked[0]

def rms(x, axis=0):
    return np.sqrt(np.mean(x**2, axis=axis))
```

```python
import matplotlib.pyplot as plt
import numpy as np
import matplotlib

# Assuming nchan and ntime are defined earlier in the code
ntime = 81  # Explicitly setting ntime as 81

# Load the results for text and audio
text = load_results(
    exp="text",
    kind="r2s",
    info=raw.info,
    nchan=nchan, 
    ntime=ntime, 
    concat=False
)

audio = load_results(
    exp="audio",
    kind="r2s",
    info=raw.info,
    nchan=nchan, 
    ntime=ntime, 
    concat=False
)

fbwav = load_results(
    exp="wavlm",
    kind="r2s",
    info=raw.info,
    nchan=nchan, 
    ntime=ntime, 
    concat=False)

wavlm = load_results(
    exp="wavLM",
    kind="r2s",
    info=raw.info,
    nchan=nchan, 
    ntime=ntime, 
    concat=False)

bert = load_results(
    exp="bert",
    kind="r2s",
    info=raw.info,
    nchan=nchan, 
    ntime=ntime, 
    concat=False)

speech = load_results(
    exp="speech",
    kind="r2s",
    info=raw.info,
    nchan=nchan, 
    ntime=ntime, 
    concat=False)

paudio = load_results(
    exp="pengi_audio",
    kind="r2s",
    info=raw.info,
    nchan=nchan, 
    ntime=ntime, 
    concat=False)

ptext = load_results(
    exp="pengi_text",
    kind="r2s",
    info=raw.info,
    nchan=nchan, 
    ntime=ntime, 
    concat=False)

# Combine results into a dictionary
all_r2s = dict(
    zip(
        ["text", "audio", "fbwav", "wavlm", "bert", "speech", "paudio"],
        [text, audio, fbwav, wavlm, bert, speech, paudio]
    )
)

colors = {
    "multimodal_text": "blue", 
    "multimodal_audio": "red", 
    "unimodal_text": "purple", 
    "unimodal_audio": "brown"
}

fig, axs = plt.subplots(1, 1, figsize=(6, 3))

# Updated x-axis values shifted by 0.2, from 0.2 to 0.8
t = np.linspace(0.2, 0.8, 61)

axs.tick_params(
    axis='both',
    which='both',
    labelsize=10,
    labelcolor="gray",
    color="gray"
)

grouped_names = {
    "multimodal_text": ["text", "speech"],
    "multimodal_audio": ["fbwav", "wavlm", "audio"],
    "unimodal_text": ["bert"],
    "unimodal_audio": ["paudio"]
}

sem_scale = 1
max_value = float('-inf')
min_value = float('inf')

for group_name, names in grouped_names.items():
    all_group_scores = []

    for name in names:
        scores_dict = all_r2s.get(name, {})
        if not scores_dict:
            print(f"No data found for {name}")
            continue

        all_scores = []
        for key, scores in scores_dict.items():
            all_scores.append(scores)

        if all_scores:
            all_scores = np.vstack(all_scores)
            try:
                all_scores = np.reshape(all_scores, (16, nchan, ntime))
            except ValueError:
                all_scores = np.reshape(all_scores, (8, nchan, ntime))
            
            all_scores = all_scores[:, :, 20:81]
            all_group_scores.append(np.mean(all_scores, axis=0))

    if all_group_scores:
        all_group_scores = np.mean(np.stack(all_group_scores), axis=0)
        for i in np.arange(61):
            all_group_scores[:, i][np.where(all_group_scores[:, i] < 0)[0]] = 0

        mean = np.nanmean(all_group_scores, axis=0)
        sem = np.std(all_group_scores, axis=0) / np.sqrt(all_group_scores.shape[0])

        color = colors[group_name]
        axs.plot(t, mean, color=color, lw=1.5, label=group_name)
        axs.fill_between(t, mean - sem * sem_scale, mean + sem * sem_scale, alpha=0.2, color=color)
        
        if max(mean + sem * sem_scale) > max_value:
            max_value = max(mean + sem * sem_scale)
            
        if min(mean - sem * sem_scale) < min_value:
            min_value = min(mean - sem * sem_scale)
    else:
        print(f"No valid scores for {group_name}")

axs.vlines(0.4, min_value, max_value, linestyle='--', lw=.8, color="gray")  # Vertical line at x=0.4 (shifted by 0.2)
axs.axhline(0.0, lw=1, color="gray", linestyle=":", zorder=0)

axs.spines["right"].set_visible(False)
axs.spines["top"].set_visible(False)
axs.legend(fontsize=8, edgecolor="none")

mf = matplotlib.ticker.ScalarFormatter(useMathText=True)
mf.set_powerlimits((-2, 2))
axs.yaxis.set_major_formatter(mf)
axs.yaxis.get_offset_text().set_size(8)

# Set custom x-axis tick labels and limit starting from 0.2 to 0.8
axs.set_xlim(0.2, 0.8)  # Ensure the x-axis starts from 0.2
axs.set_xticks(np.linspace(0.2, 0.8, 4))  # Custom ticks from 0.2 to 0.8
axs.set_xticklabels([0, 0.2, 0.4, 0.6])  # Display ticks as 0 to 0.6

axs.set_ylabel("Mean $corr$", size=10)
axs.set_xlabel("Timepoints", size=10)

plt.savefig("corr_all.png", bbox_inches="tight")
plt.savefig("corr_all.pdf", bbox_inches="tight")
plt.show()
```
