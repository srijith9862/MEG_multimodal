```python
import os
import numpy as np
from wordfreq import word_frequency, zipf_frequency, tokenize
from IPython.core.debugger import set_trace

save_dir = "features"
if not os.path.exists(save_dir):
    os.mkdir(save_dir)
```

```python
all_text = []
hp_text = []
para = ""
```

```python
narrative_text = open("./stimuli/text/lw1new.txt","r").readlines()
for eachsent in narrative_text[0].split('.'):
    all_text.append(eachsent.strip()+'.')
    for j in range(len(eachsent.strip().split())):
        #if eachsent.strip().split()[j].strip() not in [',','?','!','-','"',';']:
        hp_text.append(eachsent.strip().split()[j].strip())
print(all_text)

narrative_text = open("./stimuli/text/cable_spool_fort_new.txt","r").readlines()
for eachsent in narrative_text[0].split('.'):
    all_text.append(eachsent.strip()+'.')
    for j in range(len(eachsent.strip().split())):
        #if eachsent.strip().split()[j].strip() not in [',','?','!','-','"',';']:
        hp_text.append(eachsent.strip().split()[j].strip())
        
narrative_text = open("./stimuli/text/easy_money_new.txt","r").readlines()
for eachsent in narrative_text[0].split('.'):
    all_text.append(eachsent.strip()+'.')
    for j in range(len(eachsent.strip().split())):
        #if eachsent.strip().split()[j].strip() not in [',','?','!','-','"',';']:
        hp_text.append(eachsent.strip().split()[j].strip())
        
narrative_text = open("./stimuli/text/black_willow_new.txt","r").readlines()
for eachsent in narrative_text[0].split('.'):
    all_text.append(eachsent.strip()+'.')
    for j in range(len(eachsent.strip().split())):
        #if eachsent.strip().split()[j].strip() not in [',','?','!','-','"',';']:
        hp_text.append(eachsent.strip().split()[j].strip())
```

```python
word_frequencies = []
word_lengths = []

for i in range(len(hp_text)):
    if hp_text[i] == "+":
        word_frequencies.append(0.0)
        word_lengths.append(0)
        continue
    word_frequencies.append(zipf_frequency(hp_text[i], "en", wordlist="large"))
    word_lengths.append(len(hp_text[i]))
```

```python
for i in range(len(hp_text)):
    print("Word: {} Zipf frequency: {} Length: {}".format(hp_text[i], word_frequencies[i], word_lengths[i]))
```

```python
np.save(os.path.join(save_dir,"word_frequency"),word_frequencies)
np.save(os.path.join(save_dir,"word_length"),word_lengths)
```
