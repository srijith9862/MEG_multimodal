```python
def load_results(exp, kind, info, nchan, ntime, concat=False, reshape=True):
    files = sorted(glob.glob(f"/media/srijith/easystore/srijith/pred/meg_sub*_predictions/{exp}_joint_embeddings_predictions/"))

    
    if kind in ["corr", "r2s"]:
        res_fp = sorted(glob.glob(f"/media/srijith/easystore/srijith/pred/meg_sub*_predictions/{exp}_joint_embeddings_predictions/0_{kind}.npy"))

    print(f"Loading {exp}, found sub: {len(files)}, found files: {len(res_fp)}")

    preds = dict()
    for fp in tqdm(files):
        sub = fp.split("/")[-3].split("-")[-1]
        
        if kind in ["corr", "r2s"]:
            res_files = sorted(glob.glob(f"{fp}/0_{kind}.npy"))
            try:
                ttime=301
                if reshape:
                    preds[sub] = np.load(res_files[0]).reshape(-1, nchan, ttime).squeeze()
                else:
                    preds[sub] = np.load(res_files[0])
            except:
                print(f"Missing: {exp} - {kind}, Sub: {sub} ")
                continue
                
        else:
            res_files = sorted(glob.glob(f"{fp}/0_{kind}*"))
            results = []
            for story in range(4):
                if reshape:
                    ttime=301
                    results.append(np.load(res_files[story]).reshape(-1, nchan, ttime).squeeze())
                else:
                    results.append(np.load(res_files[story]))
            preds[sub] = np.concatenate(results)
        
    if concat:
        preds = np.concatenate(list(preds.values()))
    return preds


def as_evoked(*data, concat=False):
    evoked = []
    for datum in data:
        if concat:
            pred_epochs = np.concatenate(list(datum.values()))
            # print(pred_epochs.shape)
            evoked.append(mne.EpochsArray(pred_epochs, raw.info).average())
        else:
            pred_epochs = dict()
            for sub, preds in datum.items():
                pred_epochs[sub] = mne.EpochsArray(preds, raw.info).average()
                # print(pred_epochs.shape)
            evoked.append(pred_epochs)
        
    if len(evoked) > 1:
        return tuple(evoked)
    else:
        return evoked[0]

def rms(x, axis=0):
    return np.sqrt(np.mean(x**2, axis=axis))
```

```python
import matplotlib.pyplot as plt
import numpy as np
import matplotlib

context_audio = load_results(
    exp="context_audio",
    kind="r2s",
    info=raw.info,
    nchan=nchan, 
    ntime=ntime, 
    concat=False)
# Combine results into a dictionary
all_r2s = dict(
    zip(
        ["context_audio"],
        [context_audio]
    )
)
colors = {"context_audio":"black"}
fig, axs = plt.subplots(1, 1, figsize=(6,3))

t = np.linspace(0.0, 0.8, ntime)

axs.tick_params(
    axis='both',
    which='both',            
    labelsize=10,
    labelcolor="gray",
    color="gray"
)

names = ["context_audio"]
sem_scale = 1
max_value = 0
min_value = 0

for name in names:
    scores_dict = all_r2s[name]
    all_scores = []
    for key, scores in scores_dict.items():
        all_scores.append(scores)
    # Stack the arrays along the first axis
    all_scores = np.vstack(all_scores)
    all_scores = np.reshape(all_scores,(8,208,301))
    all_scores=np.mean(all_scores, axis=0)
    all_scores = all_scores[:,:81]
    print(all_scores.shape)
    # all_scores[np.where(all_scores<0)[0]]=0
    for i in np.arange(81):
        all_scores[:,i][np.where(all_scores[:,i]<0)[0]]=0
    mean = np.nanmean(all_scores, axis=0)
    sem = np.std(all_scores, axis=0) / np.sqrt(all_scores.shape[0])

    color = colors[name]
    axs.plot(t, mean, color=color, lw=1.5, label=name)
    axs.fill_between(t, mean - sem * sem_scale, mean + sem * sem_scale, alpha=0.2, color=color)
    
    if max(mean + sem * sem_scale) > max_value:
        max_value = max(mean + sem * sem_scale)
        
    if min(mean - sem * sem_scale) < min_value:
        min_value = min(mean - sem * sem_scale)

axs.vlines(0.2, min_value, max_value, linestyle='--', lw=.8, color="gray")
axs.axhline(0.0, lw=1, color="gray", linestyle=":", zorder=0)

axs.spines["right"].set_visible(False)
axs.spines["top"].set_visible(False)
axs.legend(fontsize=8, edgecolor="none")

mf = matplotlib.ticker.ScalarFormatter(useMathText=True)
mf.set_powerlimits((-2,2))
axs.yaxis.set_major_formatter(mf)
axs.yaxis.get_offset_text().set_size(8)

axs.set_xticks(axs.get_xticks()[::2])

axs.set_ylabel("Mean $R^2$", size=10)
axs.set_xlabel("Time (s)", size=10, loc="right", color="gray")

plt.savefig("r2s_all.png", bbox_inches="tight")
plt.savefig("r2s_all.pdf", bbox_inches="tight")
plt.show()
```
